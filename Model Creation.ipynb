```python
# for numerical computing
import numpy as np
# for dataframes
import pandas as pd
#Data Preprocessing
data=pd.read_csv("/Users/aftab/Downloads/phishcoop.csv")
data = data.drop('id',axis=1) #removing unwanted column
print(data.shape)#enables us to obtain the shape of a DataFrame.
print(data.columns)
print(data.head())
print(data.info())
data = data.drop_duplicates()
print( data.shape )
 
print(data.isnull().sum())# return the number of missing values in dataset
data=data.dropna()
print(data.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data['Result'])
plt.xlabel("-1 = Phishing Website, 1 = Non Phishing Website")
plt.title("Phishing Websites Detection")
plt.show()
```

```python
y = data.Result
# Create separate object for input features
X = data.drop('Result', axis=1)
 
# to split train and test set
from sklearn.model_selection import train_test_split
 
# Machine Learning Models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
 
# Splitting X and y into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size=0.2, random_state=0)
 
# Printing number of observations in X_train, X_test, y_train, and y_test
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
 
model1= LogisticRegression()
model2=RandomForestClassifier(random_state = 42, max_depth = 15, n_estimators = 200, min_samples_split = 2)
model4= KNeighborsClassifier(n_neighbors=7)
model5=DecisionTreeClassifier()
model6= GaussianNB()
model7=SVC()
 
model1.fit(X_train, y_train)
model2.fit(X_train, y_train)
model4.fit(X_train, y_train)
model5.fit(X_train, y_train)
model6.fit(X_train, y_train)
model7.fit(X_train, y_train)
 
# Predicting Test set results
y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)
y_pred4 = model4.predict(X_test)
y_pred5 = model5.predict(X_test)
y_pred6 = model6.predict(X_test)
y_pred7 = model7.predict(X_test)
```

```python
#For Accuracy Score
 
from sklearn.metrics import accuracy_score
acc1 = accuracy_score(y_test, y_pred1) ## get the accuracy on testing data
print("Accuracy of Logistic Regression is {:.2f}%".format(acc1*100))
 
acc2 = round(accuracy_score(y_test, y_pred2),2)## get the accuracy on testing data
print("Accuracy of RandomForestClassifier is {:.2f}%".format(acc2*100))
 
acc4 = accuracy_score(y_test, y_pred4) ## get the accuracy on testing data
print("Accuracy of KNeighborsClassifier is {:.2f}%".format(acc4*100))
 
acc5 = accuracy_score(y_test, y_pred5)  ## get the accuracy on testing data
print("Accuracy of Decision Tree is {:.2f}%".format(acc5*100))
 
acc6 = accuracy_score(y_test, y_pred6)  ## get the accuracy on testing data
print("Accuracy of GaussianNB is {:.2f}%".format(acc6*100))
 
acc7 = accuracy_score(y_test, y_pred7)  ## get the accuracy on testing data
print("Accuracy of SVC is {:.2f}%".format(acc7*100))
print(" ")
#For Precision Score
 
from sklearn.metrics import precision_score
 
prescore_1 = precision_score(y_test, y_pred1)
print("Precision_score of Logistic Regression is {:.2f}%".format(prescore_1*100))
 
prescore_2 = precision_score(y_test, y_pred2)
print("Precision_score of RandomForestClassifier is {:.2f}%".format(prescore_2*100))
 
prescore_3 = precision_score(y_test, y_pred4)
print("Precision_score of KNeighborsClassifier is {:.2f}%".format(prescore_3*100))
 
prescore_4 = precision_score(y_test, y_pred5)
print("Precision_score of Decision Tree is {:.2f}%".format(prescore_4*100))
 
prescore_5 = precision_score(y_test, y_pred6)
print("Precision_score of GaussianNB is {:.2f}%".format(prescore_5*100))
 
prescore_6 = precision_score(y_test, y_pred7)
print("Precision_score of SVC is {:.2f}%".format(prescore_6*100))
print(" ")
#For Recall Score
 
from sklearn.metrics import recall_score
 
rescore_1 = recall_score(y_test, y_pred1)
print("Recall_score of Logistic Regression is {:.2f}%".format(rescore_1*100))
 
rescore_2 = recall_score(y_test, y_pred2)
print("Recall_score of RandomForestClassifier is {:.2f}%".format(rescore_2*100))
 
rescore_3 = recall_score(y_test, y_pred4)
print("Recall_score of KNeighborsClassifier is {:.2f}%".format(rescore_3*100))
 
rescore_4 = recall_score(y_test, y_pred5)
print("Recall_score of Decision Tree is {:.2f}%".format(rescore_4*100))
 
rescore_5 = recall_score(y_test, y_pred6)
print("Recall_score of GaussianNB is {:.2f}%".format(rescore_5*100))
 
rescore_6 = recall_score(y_test, y_pred7)
print("Recall_score of SVC is {:.2f}%".format(rescore_6*100))
print(" ")
#For F1 Score
 
from sklearn.metrics import f1_score
 
f1score_1 = f1_score(y_test, y_pred1)
print("F1-Score of Logistic Regression is {:.2f}%".format(f1score_1*100))
 
f1score_2 = f1_score(y_test, y_pred2)
print("F1-Score of RandomForestClassifier is {:.2f}%".format(f1score_2*100))
 
f1score_3 = f1_score(y_test, y_pred4)
print("F1-Score of KNeighborsClassifier is {:.2f}%".format(f1score_3*100))
 
f1score_4 = f1_score(y_test, y_pred5)
print("F1-Score of Decision Tree is {:.2f}%".format(f1score_4*100))
 
f1score_5 = f1_score(y_test, y_pred6)
print("F1-Score of GaussianNB is {:.2f}%".format(f1score_5*100))
 
f1score_6 = f1_score(y_test, y_pred7)
print("F1-Score of SVC is {:.2f}%".format(f1score_6*100))
print(" ")
 
#For Confusion Matrix
 
from sklearn.metrics import confusion_matrix
 
conmat_1 = confusion_matrix(y_test, y_pred1)
print("Confusion_matrix of Logistic Regression is ",conmat_1)
 
conmat_2 = confusion_matrix(y_test, y_pred2)
print("Confusion_matrix of RandomForestClassifier is ",conmat_2)
 
conmat_3 = confusion_matrix(y_test, y_pred4)
print("Confusion_matrix of KNeighborsClassifier is ",conmat_3)
 
conmat_4 = confusion_matrix(y_test, y_pred5)
print("Confusion_matrix of Decision Tree is ",conmat_4)
 
conmat_5 = confusion_matrix(y_test, y_pred6)
print("Confusion_matrix of GaussianNB is ",conmat_5)
 
conmat_6 = confusion_matrix(y_test, y_pred7)
print("Confusion_matrix of SVC is ",conmat_6)
```
